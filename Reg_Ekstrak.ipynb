```python
from numpy import *
from scipy.interpolate import *
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from scipy.stats import *

df = pd.read_excel('input\Korelasi.xlsx')
df.tail()
```

```python
yfit_2 = polyfit(df['MQ7'], df['SPAD'], 1)[0] * df['MQ7'] + polyfit(df['MQ7'], df['SPAD'], 1)[1]
yres_2 = df['SPAD'] - yfit_2
SSresid_2 = sum(pow(yres_2,2))
SStotal_2 = len(df['SPAD']) * var(df['SPAD'])
rsq_2 = 1 - SSresid_2/SStotal_2


plot(df['MQ7'], df['SPAD'],'o')
ylabel('Uji SPAD')
xlabel('MQ7')
title("RSquare: {:.5f}".format(rsq_2), loc = 'left', fontsize=9, color='r')

xp = linspace(0, 0.2, 50)
plot(xp, polyval(polyfit(df['MQ7'], df['SPAD'], 1), xp),'r-')
plot(xp, polyval(polyfit(df['MQ7'], df['SPAD'], 2), xp),'b--')
plot(xp, polyval(polyfit(df['MQ7'], df['SPAD'], 3), xp),'m:')
```
